```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'D:\Data Analyst\HR\WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
print(df.info())
```

```python
print(df.columns)
```

```python
print(df['Attrition'].value_counts())
```

```python
#We have found that those who have low income has very less chance of being in the same company
attrition_income=df.groupby('Attrition')['MonthlyIncome'].mean()
plt.bar(attrition_income.index,attrition_income.values)
plt.xlabel('Attrition')
plt.ylabel('Income')
plt.show()
```

```python
#Those who does over time ends in Attrition
sns.countplot(data=df,x='Attrition',hue='OverTime')
plt.show()
```

```python
#Low job level employees prefer leaving the company
sns.countplot(data=df,x='Attrition',hue='JobLevel')
plt.show()
```

```python
#Salary hike does not play a significant role here
attrition_salaryhike=df.groupby('Attrition')['PercentSalaryHike'].mean()
plt.bar(attrition_salaryhike.index,attrition_salaryhike.values)
plt.show()
```

```python
#YearSinceLastPromotion gives a significant diffecence between attrition mean of years is low in 'Yes' Attrition
attrition_promotion=df.groupby('Attrition')['YearsSinceLastPromotion'].mean()
plt.bar(attrition_promotion.index,attrition_promotion.values)
plt.show()
```

```python
#In 'Yes' Attrition single has more count which tells that they are changing company due to proficitional improvement and in 'No' we could see that Married has more count whikch says that they prefer stability rather than shifting between companies
sns.countplot(data=df,x='Attrition',hue='MaritalStatus')
```
